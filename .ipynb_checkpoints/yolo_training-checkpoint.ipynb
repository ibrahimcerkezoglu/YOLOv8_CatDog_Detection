```python
from ultralytics import YOLO
import os
import cv2
import matplotlib.pyplot as plt
from pathlib import Path

import torch
print("PyTorch CUDA:", torch.cuda.is_available())
if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))
```

```python
DATA_YAML = "dataset/data.yaml"   # Roboflow'dan gelen yaml
MODEL_NAME = "yolov8n.pt"         # hazır pre-trained model

model = YOLO(MODEL_NAME)
model
```

```python
results = model.train(
    data=DATA_YAML,
    epochs=30,          # istersen sonra 50 yaparsın
    imgsz=640,
    batch=8,
    name="cats_dogs_v1",
    project="runs/train",
    patience=10
)

results
```

```python
run_dir = Path(results.save_dir)
print("Run directory:", run_dir)

img = cv2.imread(str(run_dir / "results.png"))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 6))
plt.imshow(img)
plt.axis("off")
plt.title("Eğitim Sonuç Grafikleri (Loss & mAP)")
plt.show()
```

```python
metrics = model.val(data=DATA_YAML)
metrics.results_dict
```

```python
test_images_dir = "dataset/test/images"
sample_image = os.path.join(test_images_dir, os.listdir(test_images_dir)[0])
sample_image
```

```python
results_infer = model(sample_image)

for r in results_infer:
    im = r.plot()  # BGR görüntü
    im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(6, 6))
    plt.imshow(im_rgb)
    plt.axis("off")
    plt.title("Örnek Tahmin (Kedi/Köpek)")
    plt.show()
```

```python
import shutil

best_model_src = run_dir / "weights" / "best.pt"
best_model_dst = "best.pt"

shutil.copy(best_model_src, best_model_dst)
print("best.pt kopyalandı ->", best_model_dst)
```
